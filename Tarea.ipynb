```
df = pd.read_csv("/content/Casos1.csv")
dist=df["Fecha de diagnóstico"]
cont = []
for i in range(0,len(dist)):
    cont.append(i+1)

df["Contador"]=cont
```

```
columns=df[["Fecha de diagnóstico", "Contador"]]
dat=pd.to_datetime(columns["Fecha de diagnóstico"],  dayfirst=True)
print(columns["Fecha de diagnóstico"][:1])
```

```
#columns['Fecha'] = dat
#columns.head(10)
#print(len(columns["Fecha de diagnóstico"]))

var=columns["Fecha de diagnóstico"][0]
fechas=[]
len_dat=len(columns["Fecha de diagnóstico"])
for i in range(len_dat):
  # print(columns["Fecha de diagnóstico"][i])
  if(var!=columns["Fecha de diagnóstico"][i]):
    fechas.append([columns["Fecha de diagnóstico"][i-1],columns["Contador"][i-1]])
    var=columns["Fecha de diagnóstico"][i]
fechas.append([columns["Fecha de diagnóstico"][len_dat-1],columns["Contador"][len_dat-1]])

print("Fecha\t\tCaso")
for i in fechas:
  print(i[0],"\t",i[1])

```

```
fechas=pd.DataFrame(fechas)
fechas=fechas.rename(columns={0:"Fechas",1:"Casos"})

Count=[]
for i in range(len(fechas)):
  Count.append(i+1)
#print(Count)
fechas["Días"]=Count

Datos=fechas[["Días","Casos"]]
Datos.head(38)
```
